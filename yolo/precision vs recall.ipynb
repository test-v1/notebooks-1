```python
import re
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)
np.set_printoptions(threshold=np.nan)

def pvr(out_file):

    F = 0.5
    lines = open(out_file, 'r').readlines()
    r = re.compile("^/data1/yolo/model")
    lines = filter(r.match, lines)
    lines = [ [int(l.split(' ')[0].split('.')[-2].split('_')[1]), float(l.split(' ')[1].strip()), float(l.split(' ')[2].strip())] for l in lines ]
    lines = np.array(lines)
    # lines2 = np.hstack((lines2, np.reshape(lines2[:,1]*lines2[:,1]*lines2[:,2]*2, (-1, 1))))
    lines = np.hstack((lines, np.reshape((1+F*F)*(lines[:,1]*lines[:,2])/(F*F*lines[:,1]+lines[:,2]), (-1, 1))))

    lines = lines[np.argsort(lines[:, 0])]
    df = pd.DataFrame(data=lines[:,1:], index=lines[:,0], columns=['precision', 'recall', 'f0.5'])
    df.plot.line()
    return df
```
