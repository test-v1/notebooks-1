```python
import json
import sys
from os import path
from os import listdir

videos = listdir('/data/yolo/detector/web/imgs')

failure_starts_detected = {}

def failure_start(frames):    
    for frm in frames:
        if frm['score'] > 0.2:
            return int(frm['img'].replace('.jpg', ''))
    return -1

for vid in videos:
    with open('/data/yolo/detector/web/imgs/%s/detections.json' % (vid)) as f:
        failure_starts_detected[vid] = failure_start(json.load(f))
```
